```python
import nltk
from nltk.chat.util import Chat, reflections

# Define the pairs of patterns and responses related to student information and professional interview questions
pairs = [
    # Student Information
    [
        r"tell me about yourself",
        ["I am Neha. I am studying B.Tech in Computer Science and Engineering with a focus on AI & DS at AKS University, Satna. My major is AI & ML. I enjoy solving problems on coding platforms."]
    ],
    [
        r"what is your name?",
        ["My name is Neha."]
    ],
    [
        r"what are you studying?",
        ["I am studying B.Tech in Computer Science and Engineering with a focus on AI & DS at AKS University, Satna."]
    ],
    [
        r"which institution are you studying at?",
        ["I am studying at AKS University, Satna."]
    ],
    [
        r"what is your major?",
        ["My major is AI & ML."]
    ],
    [
        r"where do you live?",
        ["I live in Satna."]
    ],
    [
        r"what year are you in?",
        ["I am currently in my B.Tech program."]
    ],
    [
        r"what are your hobbies?",
        ["I enjoy solving problems on coding platforms."]
    ],
    [
        r"what are your academic interests?",
        ["I am interested in AI & ML."]
    ],

    # Professional Interview Questions
    [
        r"what are your strengths?",
        ["My strengths include problem-solving skills, analytical thinking, and the ability to quickly adapt to new technologies."]
    ],
    [
        r"what are your weaknesses?",
        ["One of my weaknesses is that I sometimes spend too much time on details. However, I am working on improving my time management skills."]
    ],
    [
        r"why should we hire you?",
        ["You should hire me because I have a strong background in AI & ML and have demonstrated my problem-solving abilities through various coding challenges. I am eager to apply my skills and learn from the team."]
    ],
    [
        r"where do you see yourself in five years?",
        ["In five years, I see myself as a skilled AI/ML specialist, working on innovative projects and contributing to advancements in technology. I aim to grow professionally and take on leadership roles."]
    ],
    [
        r"why do you want to work here?",
        ["I am interested in working here because of the company's focus on cutting-edge technology and its commitment to innovation. I am excited about the opportunity to contribute to impactful projects and further develop my skills."]
    ],
    [
        r"can you describe a challenging project you worked on?",
        ["One challenging project I worked on involved developing a machine learning model to predict customer behavior. It required extensive data preprocessing and model tuning. Through this experience, I enhanced my skills in data analysis and machine learning algorithms."]
    ],
    [
        r"how do you handle stress and pressure?",
        ["I handle stress and pressure by staying organized, setting clear priorities, and maintaining a positive attitude. I also find that taking short breaks and practicing mindfulness helps me stay focused and manage stress effectively."]
    ],
    [
        r"what motivates you?",
        ["I am motivated by the opportunity to solve complex problems and make a positive impact through technology. The challenge of learning new skills and achieving goals also drives me to perform at my best."]
    ],
    [
        r"bye|goodbye|see you",
        ["Goodbye! Have a great day!", "See you later!"]
    ]
]

# Create the chatbot
chatbot = Chat(pairs, reflections)

# Function to start the conversation
def start_chatbot():
    print("Chatbot: Hi! I'm your student and professional interview chatbot. Type 'bye' to exit.")
    while True:
        user_input = input("You: ").lower()
        if user_input == 'bye':
            print("Chatbot: Goodbye! Have a great day!")
            break
        response = chatbot.respond(user_input)
        if response:
            print(f"Chatbot: {response}")
        else:
            print("Chatbot: I'm sorry, I didn't understand that.")

# Start the chatbot
if __name__ == "__main__":
    start_chatbot()
```
